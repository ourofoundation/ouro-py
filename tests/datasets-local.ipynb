```python
import os
from ouro import Ouro
import pandas as pd
import dotenv

dotenv.load_dotenv()

ouro = Ouro(
    api_key=os.environ.get("OURO_API_KEY"),
    base_url=os.environ.get("OURO_BACKEND_URL")
)
```

```python
df = pd.DataFrame({"a": [1, 2, 3], "b": [4, 5, 6]})

dataset = ouro.datasets.create(
    name="test dataset",
    visibility="public",
    description="test dataset",
    data=df
)

dataset
```
