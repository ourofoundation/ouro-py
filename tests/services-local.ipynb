```python
generation = ouro.routes.use(
   "mmoderwell/post-chemeleon-generate-text",
   body={
"n_atoms": 6,
"n_samples": 1,
"text_input": "A crystal structure of FeMnO4 with orthorhombic symmetry"
    }
)
```

```python
import os
from ouro import Ouro
import dotenv
import requests
import numpy as np
from typing import Optional, Tuple, Union
from pymatgen.core import Structure, Composition
import zipfile
import io


file_id = generation["file"]["id"]

file = ouro.files.retrieve(file_id)
file_data = file.read_data()
print(file_data.url)
url = file_data.url
response = requests.get(url)
zip_bytes = response.content
cif_output_dir = "cif_output"

os.makedirs(cif_output_dir, exist_ok=True)


with zipfile.ZipFile(io.BytesIO(zip_bytes)) as zf:
    cif_filenames = [name for name in zf.namelist() if name.lower().endswith(".cif")]
    if not cif_filenames:
        raise ValueError("No CIF file found in the ZIP archive.")

    cif_filename = cif_filenames[0]
    cif_bytes = zf.read(cif_filename)

print(f"Extracted CIF file: {cif_filename}")
print(f"raw: {cif_bytes}")
print("Converting to pymatgen structure...")

cif_str = cif_bytes.decode("utf-8")
structure = Structure.from_str(cif_str, fmt="cif")
print(f"Created structure with formula {structure.formula}")

output_cif_path = os.path.join(cif_output_dir, os.path.basename(cif_filename))
with open(output_cif_path, "w", encoding="utf-8") as f:
    f.write(cif_str)

print(f"Wrote CIF to {output_cif_path}")

written_file = ouro.files.create(
    name=f'{generation["file"]["name"]}-cif',
    description=f'{generation["file"]["description"]}-cif',
    visibility="public",
    file_path=output_cif_path,
)


print(written_file)
```
