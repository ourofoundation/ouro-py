```python
import os
from ouro import Ouro
import pandas as pd
import dotenv

dotenv.load_dotenv()

ouro = Ouro(
    api_key=os.environ.get("OURO_API_KEY"),
    base_url=os.environ.get("OURO_BACKEND_URL")
)
```

```python
file = ouro.files.create(
    name="cif file from API, updated",
    description="Test file",
    visibility="public",
    file_path="./Fe.cif",
)
print(file)
```
